```python
import pandas as pd
import numpy as np
# import ciso8601
from datetime import datetime
from tqdm import tqdm 
import pickle
import random
import os
```

```python
time_interval = '18_39'
test_file = '../processing_data/pickle/trimming/12_with_gap_3/' + time_interval + '/final_test.pickle'
with open(test_file,'rb') as f:
    test = pickle.load(f)
```

```python
amount_list = []
for i in range(0, len(test[0])):
    num = 0
    for j in range(0, len(test[0][i])):
        for element in range(0, len(test[0][i][j])):
            num += 1;
    amount_list.append(num)
new_list = sorted(amount_list)
```

```python
index_list = []
for x in range(len(new_list) - 11, len(new_list)):
    for y in range(0, len(amount_list)):
        if amount_list[y] == new_list[x] and y not in index_list:
            index_list.append(y)
```

```python
print(index_list)
```

```python
test0 = []
test1 = []
test2 = []
test3 = []
for element in index_list:
    test0.append(test[0][element])
    test1.append(test[1][element])
    test2.append(test[2][element])
    test3.append(test[3][element])
for x in range(0, len(test3)):
    for y in range(0, len(test3[x])):
        for z in range(0, len(test3[x][y])):
            test3[x][y][z] = float(test3[x][y][z])
```

```python
final_test = np.array([test0, test1, test2, test3])
with open('../processing_data/pickle/showing/final_test.pickle', 'wb') as b:
    pickle.dump(final_test,b)
```
