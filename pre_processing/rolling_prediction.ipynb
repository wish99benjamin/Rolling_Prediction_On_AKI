```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
time_interval = '29_39'
train_file = '../processing_data/pickle/unbalanced/' + time_interval + '/final_train.pickle'
val_file = '../processing_data/pickle/unbalanced/' + time_interval + '/final_val.pickle'
test_file = '../processing_data/pickle/unbalanced/' + time_interval + '/final_test.pickle'
with open(train_file,'rb') as f:
    train = pickle.load(f)
    
with open(val_file,'rb') as f:
    val = pickle.load(f)
    
with open(test_file,'rb') as f:
    test = pickle.load(f)
```

```python
print(train[0][0])
```

```python
# 把分開的3個檔案收集起來
zero = train[0] + val[0] + test[0]
one = train[1] + val[1] + test[1]
two = train[2] + val[2] + test[2]
three = train[3] + val[3] + test[3]
new_zero = []
new_one = []
new_two = []
new_three = []
```

```python
print(len(zero))
```

# 篩選有最多紀錄的30個(0跟1各15個)

```python
# 選出有最多特徵的30個
amount_list = []
for i in range(0, len(zero)):
    num = 0
    for j in range(0, len(zero[i])):
        for element in range(0, len(zero[i][j])):
            num += 1;
    amount_list.append(num)
new_list = sorted(amount_list)
```

```python
index_list = []
x = len(new_list) - 1
num_of_positive = 0
num_of_negative = 0

while len(index_list) <= 30: 
    for y in range(0, len(amount_list)):
        if amount_list[y] == new_list[x] and y not in index_list:
            print(one[y])
            if one[y] == 1 and num_of_positive <= 15:
                index_list.append(y)
                num_of_positive += 1
            elif one[y] == 0 and num_of_negative <= 15:
                index_list.append(y)
                num_of_negative += 1
    x -= 1
```

```python
temp0 = []
temp1 = []
temp2 = []
temp3 = []
for element in index_list:
    temp0.append(zero[element])
    temp1.append(one[element])
    temp2.append(two[element])
    temp3.append(three[element])
```

```python
print(temp1)
```

```python
start = 0
end = start + 11
for index in range(0, len(temp0)):
    for num in range(0, len(temp0[index])):
        start = num
        end = start + 11
        if end > len(temp0[index]) - 6:
            break
        new_zero.append(temp0[index][start:end])
        new_one.append(temp1[index])
        new_two.append(temp2[index][start:end])
        new_three.append(temp3[index][start:end])
```

# 沒有篩選(固定長度)

```python
start = 0
end = start + 11
for index in range(0, len(zero)):
    for num in range(0, len(zero[index])):
        start = num
        end = num + 11
        if end > len(zero[index]) - 6:
            break
        new_zero.append(zero[index][start:end])
        new_one.append(one[index])
        new_two.append(two[index][start:end])
        new_three.append(three[index][start:end])
```

# 沒有篩選(越來越長)

```python
start = 0
end = start + 11
for index in range(0, len(zero)):
    for num in range(0, len(zero[index])):
        start = 0
        end = num + 11
        if end > len(zero[index]) - 6:
            break
        new_zero.append(zero[index][start:end])
        new_one.append(one[index])
        new_two.append(two[index][start:end])
        new_three.append(three[index][start:end])
```

# 生成pickle

```python
final = np.array([new_zero, new_one, new_two, new_three])
final = final.tolist()
```

```python
print(len(final[0]))
```

```python
if not os.path.exists('../processing_data/pickle/rolling_prediction/' + time_interval):
    os.mkdir('../processing_data/pickle/rolling_prediction/' + time_interval)
with open('../processing_data/pickle/rolling_prediction/' + time_interval + '/rp_vlength.pickle', 'wb') as b:
    pickle.dump(final,b)
```
