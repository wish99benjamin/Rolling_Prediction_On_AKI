```python
def normalize(input,mini,maxi):
#     print(input)
    if input == "≧90":
        input = float(90)
    elif input == "Error":
        return False
    elif input == "/" or input == '\\':
        return False
    elif '<' in str(input) :
        input = float(input[1:]) - 0.01
    elif '>' in str(input) :
        input = float(input[1:]) + 0.01
    else:
        input = float(input)
    return (input-float(mini))/(float(maxi)-float(mini))
```

```python
def list_maker(num, mode):
    product = []
    for i in range(num):
        if mode ==  0:
            product.append([])
        else: 
            product.append(0)
    return product
```

```python
for index in range(0, feature_amount):
    feature_num = index + 11
    time = 0
    print(feature_list[index])
    try:
        df = pd.read_csv(source_dir + feature_list[index] + ".csv", header = None, encoding = 'Big5', engine = 'python')
    except:
        df = pd.read_csv(source_dir + feature_list[index] + ".csv", header = None, encoding = 'utf-8', engine = 'python')
    df.drop([4,3,1], axis = 1, inplace = True)
    df = df.rename(columns = {0: 'Index',2: 'Timestamp', 5 : 'Value'})
    #print(df)
    df_time = []
    for c in df['Timestamp']:
        df_time.append(ciso8601.parse_datetime(c[:-4]).timestamp())
    df = df.drop(columns = 'Timestamp')
    df['Time'] = df_time
    df['Index'] = df['Index'].astype(int)
    df = df.sort_values(by='Index')
    df = df.reset_index(drop=True)
    
    mini = df['Value'].min()
    maxi = df['Value'].max()
    
    for unit in tqdm(range(0,len(patientList))):
        st = patientList['start_time'][unit]
        et = patientList['end_time'][unit]
        interval = int((et - st)/3600)
        time_check = list_maker(interval, 1)
        temp = df['Index'].isin([patientList['Index'][unit]])
        cur = temp[temp == True].index.tolist()
        for i in range(len(cur)):
            t = df['Time'][cur[i]] + 28800
            if t >= st and t <= et:
                t = t - st
                t /= 60
                t = int(t // 60)
                if t < interval:
                    value = (df['Value'][cur[i]]-float(mini))/(float(maxi)-float(mini))
                    if value != False:
                        if time_check[t] == 0:
                            time += 1
                            big_dim1[unit][t].extend([feature_num])
                            value = float(value)
                            big_dim4[unit][t].extend([value])
                        time_check[t] +=1
    print(time)
```
